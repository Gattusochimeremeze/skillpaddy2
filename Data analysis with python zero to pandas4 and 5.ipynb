### Customizing the xaxis

### AXIS LABELS

```python
plt.plot(years, yield_apples)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
```

### Plotting multiple lines

```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896]
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
```

### CHART TITLE AND LEGEND

```python
plt.plot(years, apples)
plt.plot(years, oranges)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop yields in kanto")
plt.legend(['apples', 'oranges']);
```

### LINE MARKERS

```python
plt.plot(years, apples, marker='o')
plt.plot(years, oranges, marker='x')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop yields in kanto")
plt.legend(['apples', 'oranges']);
```

#### STYLING LINE AND MARKERS

```python
plt.plot(years, apples, marker='o', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, oranges, marker='x', c='r', ls='--', lw=3, ms=10, alpha=.5)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop yields in kanto")
plt.legend(['apples', 'oranges']);
```

```python
#fmt argument provides a shorthand for specifying the line style, marker, and line color..it can be as a third argument to plt.plot
#fmt = '[marker][line][color]'
```

```python
plt.plot(years, apples, 's-b')
plt.plot(years, oranges, 'o--r')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop yields in kanto")
plt.legend(['apples', 'oranges']);
```

```python
#if no line style is soecified in fmt only markers are drawn
```

```python
plt.plot(years, oranges, 'or')
plt.title("Yield of oranges (tons per hectare)");
```

### CHANGING THE FIGURE SIZE

```python
#we use the plt.figure function to change the size of the figure
```

```python
plt.figure(figsize=(12, 6))

plt.plot(years, oranges, 'or')
plt.title("yield of oranges (tons per hectare)");
```

## BAR CHART

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
years = range(2000, 2006)
apples = [0.35, 0.6, 0.9, 0.8, 0.65, 0.8]
oranges = [0.4, 0.8, 0.9, 0.7, 0.6, 0.8]
```

```python
plt.plot(years, oranges);
```

```python
plt.bar(years, oranges)
```

```python
plt.bar(years, oranges);
plt.plot(years, oranges, 'o--r');
plt.title("yield of oranges")
```

```python
plt.bar(years, apples)
plt.bar(years, oranges, bottom=apples);
```
